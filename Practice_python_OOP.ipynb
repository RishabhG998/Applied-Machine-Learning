@staticmethod - Static methods, much like class methods, are methods that are bound to a class rather than its object, they do not require a class instance creation. So, they are not dependent on the state of the object.

```python
# create a class (create blueprint)
class SoftwareEngineer:

  # class attribute
  alias = "Magician"

  # define class elements
  def __init__(self,name, age, level, salary):
    # instance attributes
    self.name = name
    self.age = age
    self.level = level
    self.salary = salary

  # instance method
  def code(self):
    print(f'{self.name} is writting the code')

  # instance method with parameters
  def code_in_language(self, language):
    print(f'{self.name} is writting the code in {language}')

  # return from instance method
  def info(self):
    info = f'{self.name} is {self.age} years old'
    return info

  # dunder instance method
  def __str__(self):
    info = f'{self.name} is {self.age} years old'
    return info

  def __eq__(self, other):
    return self.name == other.name  and self.age == other.age

  @staticmethod
  def calc_salary(age):
    if age <= 25:
      return 5000
    elif age < 30:
      return 7000
    return 9000

# instance
se1 = SoftwareEngineer("max", 25, "entry", 7000)
se2 = SoftwareEngineer("max", 25, "entry", 7000)

print(se1.info())
print(se1)

print(se1==se2)
# print(se1.calc_salary())
# it returns false normally because it checks memory location


print(SoftwareEngineer.calc_salary(27))
```

### Inheritance

The process by which one class takes the attributes and methods of the another. The newly formed class is called child class and the class from which the child class is derived is called parent class.

Child class inherit all attributes and methods from parent class but can also overrite and extend methods/attributes that are unique to themselves

Child class inherits all attributes and functions from parent class

```python
# class Employee is the parent class
class Employee:

  def __init__(self, name, age):
    self.name = name
    self.age = age

# SoftwareEngineer is the child class of class Employee
class SoftwareEngineer(Employee):
  pass

# Designer is the child class of class Employee
class Designer(Employee):
  pass

se = SoftwareEngineer("Max", 25)
d = Designer("Sam", 29)

print(se.name, se.age)
print(d.name, d.age)
```

Adding a function - child class inherits the function

### Polymorphism

"Many shapes"

We write a code that works on superclass then it will work on a subclass type also.

Gives a way to use a class exactly like it's parents but the child class keeps its own methods as well.

```python
employees = [SoftwareEngineer("Max", 25, 5000, 'Junior'),
             SoftwareEngineer("Paul", 24, 4000, 'Junior'),
             Designer("Sam", 29, 7000)]

def motivate_employees(employee):
  for employee in employees:
   print(employee.working())

motivate_employees(employees)
```

```python
class SoftwareEngineer:

  def __init__(self, name, age, level):
    self.name = name
    self.age = age
    self.__salary = None
    self._number_of_bugs_solved = 0
    self.level = level

  def bugs(self):
    self._number_of_bugs_solved += 1

  # getter
  def get_salary(self):
    return self.__salary

  # setter
  def set_salary(self, value):

    if value < 1000:
      self.__salary = 1000
    elif value > 20000:
      self.__salary = 20000
    self.__salary = value


se = SoftwareEngineer("Max", 25, "Junior")
print(se.name, se.level)

```

Adding base value and salary calculation

```python
class SoftwareEngineer:

  def __init__(self, name, age, level):
    self.name = name
    self.age = age
    # private attribute
    self.__salary = None
    self._number_of_bugs_solved = 0
    self.level = level

  def bugs(self):
    self._number_of_bugs_solved += 1

  # getter
  def get_salary(self):
    return self.__salary

  # setter
  def set_salary(self, base_value):

    self.__salary = self._calculation_of_salary(base_value)

  # private function
  def _calculation_of_salary(self, base_value):

    if self._number_of_bugs_solved < 5:
      return base_value
    elif self._number_of_bugs_solved < 20:
      return base_value*2
    return base_value * 3

se = SoftwareEngineer("Max", 25, "Junior")
print(se.name, se.level)


for i in range(15):
  se.bugs()

se.set_salary(6000)
print(se.get_salary())
```

### Abstraction

An extension of Encapsulation, each object should only expose a hihg-level mechanism for using it

This mechanism should hide internal implementation details. It should only reveal operations relevant for the other objects

### Property
